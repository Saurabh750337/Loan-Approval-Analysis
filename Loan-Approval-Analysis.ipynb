```python
# Set the figure size
plt.figure(figsize=(12, 8))

# Box Plot: Loan Amount vs Gender
plt.subplot(2, 2, 1)
sns.boxplot(x=data['Gender'], y=data['LoanAmount'], palette='Set2')
plt.title("Loan Amount vs Gender")
plt.xlabel("Gender")
plt.ylabel("Loan Amount")

# Violin Plot: Applicant Income vs Education
plt.subplot(2, 2, 2)
sns.violinplot(x=data['Education'], y=data['ApplicantIncome'], palette='Set3')
plt.title("Applicant Income vs Education")
plt.xlabel("Education")
plt.ylabel("Applicant Income")



# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation matrix for numeric variables
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = data[numeric_columns].corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    correlation_matrix, 
    annot=True,       
    cmap='coolwarm',
    fmt='.2f',        
    linewidths=0.5     
)
plt.title("Correlation Heatmap")
plt.show()
```

```python
counts = data.groupby(['Education', 'Gender']).size()

# Plot the stacked bar chart
counts.plot(kind='bar', stacked=True, figsize=(10, 6), color=['lightblue', 'lightcoral'])

plt.title("Distribution of Gender across Education Levels")
plt.xlabel("Education Level")
plt.ylabel("Count")
plt.xticks(rotation=0) 
plt.tight_layout()
plt.show()
```
